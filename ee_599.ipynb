```python
# Normal GCD
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

a = 48
b = 18
gcd_result = gcd(max(a, b), min(a, b))  # Swap a and b if a < b
print(f"Normal GCD({a}, {b}) = {gcd_result}")

def binary_gcd(a, b):
    if a == b:
        return a

    if a == 0:
        return b

    if b == 0:
        return a

    if a % 2 == 0 and b % 2 == 0:
        # Both a and b are even, gcd(a, b) = 2 * gcd(a/2, b/2)
        return 2 * binary_gcd(a // 2, b // 2)
    elif a % 2 == 0:
        # a is even and b is odd, gcd(a, b) = gcd(a/2, b)
        return binary_gcd(a // 2, b)
    elif b % 2 == 0:
        # a is odd and b is even, gcd(a, b) = gcd(a, b/2)
        return binary_gcd(a, b // 2)
    elif a > b:
        # Both a and b are odd, and a > b, gcd(a, b) = gcd(a - b, b)
        return binary_gcd(a - b, b)
    else:
        # Both a and b are odd, and a < b, gcd(a, b) = gcd(b - a, a)
        return binary_gcd(b - a, a)

# Example usage:
a = 48
b = 18
result = binary_gcd(a, b)
print(f'GCD({a}, {b}) = {result}')
```

Why?

Normal gcd takes time for large numbers due to number of iterations


Binary GCD: More complex to understand but highly efficient, making it suitable for finding the GCD of large numbers

# **Bezout's Identity:**
Bezout's Identity states that for any two non-negative integers a and b, there exist integers x and y such that:

ax + by = gcd(a, b)


Algorithm:


Algorithm:

Initialize variables:
Let a and b be the two input integers.

Set x1, x2, y1, and y2 to 1, 0, 0, and 1 respectively.

While b is not equal to 0:
Calculate q as the integer division of a by b.

Update a to b and b to the remainder of a divided by b.

Update x1 to x2, x2 to x1 - q * x2, y1 to y2, and y2 to y1 - q * y2.

Once b becomes 0, the GCD of the original a and b is stored in a, and the Bézout coefficients x1 and y1 are the solutions to the equation ax + by = GCD(a, b).

```python
#Bezout's Identity:
def bezout_identity_gcd(a, b):
    if b == 0:
        return a, 1, 0
    else:
        gcd, x, y = bezout_identity_gcd(b, a % b)
        x, y = y, x - (a // b) * y
        return gcd, x, y

a = 48
b = 18
gcd, x, y = bezout_identity_gcd(a, b)

print(f'GCD({a}, {b}) = {gcd}')
print(f'x = {x}, y = {y}')
```

```python
#Extended Euclidean Algorithm:

def extended_gcd(a, b):
    if a == 0:
        return b, 0, 1
    else:
        gcd, x, y = extended_gcd(b % a, a)
        x, y = y - (b // a) * x, x
        return gcd, x, y

a = 48
b = 18
gcd, x, y = extended_gcd(a, b)

print(f'GCD({a}, {b}) = {gcd}')
print(f'x = {x}, y = {y}')
```
